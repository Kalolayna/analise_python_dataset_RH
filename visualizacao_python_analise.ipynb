```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
# Carregando Dataset Funcion√°rios 
df = pd.read_csv('HRDataset_v14.csv')
```

```python
# Estat√≠sticas descritivas b√°sicas
desc_stats = df.describe()
desc_stats
```

```python
# Calcula a idade dos funcion√°rios subtraindo a data de nascimento da data de hoje
df['DOB'] = pd.to_datetime(df['DOB'])
hoje = datetime.datetime.today()
df['Age'] = ((hoje - df['DOB']).dt.days / 365.25).astype(int)
```

```python
df.head()
```

```python
# Informa√ß√µes do Dataset
df.info
```

```python
# Media sal√°rio e satisfa√ß√£o funcion√°rios 
media_salario = df['Salary'].mean()
media_satisfacao_funcionario = df['EmpSatisfaction'].mean()
print("M√©dia de sal√°rio:", media_salario)
print("M√©dia de satisfa√ß√£o:", media_satisfacao_funcionario)
```

```python
# Histograma com sal√°rios mais populares entre os funcion√°rios
plt.figure
sns.displot(data= df, x= "Salary", bins= 5)
plt.title('Faixas de Sal√°rio Populares')
plt.ylabel('Frequ√™ncia')
plt.xlabel('Valor Sal√°rio')
plt.show()
```

```python
# Violino com sal√°rios, discriminados por g√™nero 
plt.figure
sns.violinplot(data=df, x='Sex', y='Salary', hue='Sex', palette='Set3', legend=False)
plt.title('Sal√°rios por G√™nero')
plt.xlabel('G√™nero')
plt.ylabel('Sal√°rio')
plt.show()
```

```python
# Convers√£o da coluna DateofHire para tipo datetime
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DOB'] = pd.to_datetime(df['DOB'])
```

```python
# Gr√°fico de s√©rie temporal de data de contrata√ß√£o
plt.figure
sns.lineplot(data=df, x='DateofHire', y=df.index, errorbar=None)
plt.title('S√©rie Temporal de Data de Contrata√ß√£o')
plt.xlabel('Data de Contrata√ß√£o')
plt.ylabel('√çndice')
plt.show()
```

```python
# Gr√°fico de s√©rie temporal de data de contrata√ß√£o
plt.figure
sns.lineplot(data=df, x='DateofHire', y='EmpSatisfaction', errorbar=None)
plt.title('S√©rie Temporal de Satisfa√ß√£o dos funcion√°rios')
plt.xlabel('Data de avalia√ß√£o')
plt.ylabel('√çndice')
plt.show()
```

```python
# Forma de contrata√ß√£o
plt.figure
sns.countplot(data=df, y='RecruitmentSource', palette='Set3')
plt.ylabel('Fonte de Recrutamento')
plt.xlabel('Contagem')
plt.title('Contagem de Fontes de Recrutamento')
plt.show()
```

```python
# Distribui√ß√£o salarial entre os departamentos 
plt.figure
sns.boxplot(data=df, x='Department', y='Salary')
plt.xlabel('Departamentos')
plt.ylabel('Sal√°rios')
plt.title('Distribui√ß√£o Salarial')
plt.xticks(rotation=45)
plt.show()
```
