```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import datetime
```

```python
# Carregando Dataset Funcionários 
df = pd.read_csv('HRDataset_v14.csv')
```

```python
# Estatísticas descritivas básicas
desc_stats = df.describe()
desc_stats
```

```python
# Calcula a idade dos funcionários subtraindo a data de nascimento da data de hoje
df['DOB'] = pd.to_datetime(df['DOB'])
hoje = datetime.datetime.today()
df['Age'] = ((hoje - df['DOB']).dt.days / 365.25).astype(int)
```

```python
df.head()
```

```python
# Informações do Dataset
df.info
```

```python
# Media salário e satisfação funcionários 
media_salario = df['Salary'].mean()
media_satisfacao_funcionario = df['EmpSatisfaction'].mean()
print("Média de salário:", media_salario)
print("Média de satisfação:", media_satisfacao_funcionario)
```

```python
# Histograma com salários mais populares entre os funcionários
plt.figure
sns.displot(data= df, x= "Salary", bins= 5)
plt.title('Faixas de Salário Populares')
plt.ylabel('Frequência')
plt.xlabel('Valor Salário')
plt.show()
```

```python
# Violino com salários, discriminados por gênero 
plt.figure
sns.violinplot(data=df, x='Sex', y='Salary', hue='Sex', palette='Set3', legend=False)
plt.title('Salários por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Salário')
plt.show()
```

```python
# Conversão da coluna DateofHire para tipo datetime
df['DateofHire'] = pd.to_datetime(df['DateofHire'])
df['DOB'] = pd.to_datetime(df['DOB'])
```

```python
# Gráfico de série temporal de data de contratação
plt.figure
sns.lineplot(data=df, x='DateofHire', y=df.index, errorbar=None)
plt.title('Série Temporal de Data de Contratação')
plt.xlabel('Data de Contratação')
plt.ylabel('Índice')
plt.show()
```

```python
# Gráfico de série temporal de data de contratação
plt.figure
sns.lineplot(data=df, x='DateofHire', y='EmpSatisfaction', errorbar=None)
plt.title('Série Temporal de Satisfação dos funcionários')
plt.xlabel('Data de avaliação')
plt.ylabel('Índice')
plt.show()
```

```python
# Forma de contratação
plt.figure
sns.countplot(data=df, y='RecruitmentSource', palette='Set3')
plt.ylabel('Fonte de Recrutamento')
plt.xlabel('Contagem')
plt.title('Contagem de Fontes de Recrutamento')
plt.show()
```

```python
# Distribuição salarial entre os departamentos 
plt.figure
sns.boxplot(data=df, x='Department', y='Salary')
plt.xlabel('Departamentos')
plt.ylabel('Salários')
plt.title('Distribuição Salarial')
plt.xticks(rotation=45)
plt.show()
```
